```python
from selenium import webdriver
import time
```

```python
url = 'https://news.naver.com/main/ranking/read.nhn?mode=LSD&mid=shm&sid1=001&oid=025&aid=0003088732&rankingType=RANKING'
```

```python
driver =webdriver.Chrome('C:/Users/SANGMIN/Downloads/chromedriver_win32(1)/chromedriver.exe')
driver.implicitly_wait(30)
driver.get(url)

time.sleep(1)
btn_add = driver.find_element_by_css_selector('span.u_cbox_in_view_comment')
btn_add.click()

while True:
    try :
        btn_add1 = driver.find_element_by_css_selector('a.u_cbox_btn_more')
        btn_add1.click()
    except:
        break
    
news_title = driver.find_elements_by_css_selector('div.article_info > h3 > a')
print("기사 제목 : " + news_title[0].text)

news_comments = driver.find_elements_by_css_selector('span.u_cbox_contents')
i = 1
for comment in news_comments:
    print(i,"번째 댓글 : ",comment.text)
    i += 1
```

```python
def news_crawling(url, wait_time):
    driver = webdriver.Chrome('C:/Users/SANGMIN/Downloads/chromedriver_win32(1)/chromedriver.exe')
    driver.implicitly_wait(wait_time)
    driver.get(url)
    
    time.sleep(1)
    btn_add = driver.find_element_by_css_selector('span.u_cbox_in_view_comment')
    btn_add.click()

    while True:
        try :
            btn_add1 = driver.find_element_by_css_selector('a.u_cbox_btn_more')
            btn_add1.click()
        except:
            break
            
    html = driver.page_source
    
    from bs4 import BeautifulSoup
    soup = BeautifulSoup(html,'lxml')
    
    dates = soup.select('span.u_cbox_date')
    dates = [ date.text for date in dates]
    
    comments = soup.select('span.u_cbox_contents')
    comments = [comment.text for comment in comments]
    
    data = list(zip(dates,comments))
    driver.quit()
    return data
```

```python
url = 'https://news.naver.com/main/ranking/read.nhn?mode=LSD&mid=shm&sid1=001&oid=025&aid=0003088732&rankingType=RANKING'

news_crawling(url,30)
```
